### About the dataset

### Content
The dataset contains transactions made by credit cards in September 2013 by European cardholders.
This dataset presents transactions that occurred in two days, where we have 492 frauds out of 284,807 transactions. The dataset is highly unbalanced, the positive class (frauds) account for 0.172% of all transactions.

It contains only numerical input variables which are the result of a PCA transformation. Unfortunately, due to confidentiality issues, we cannot provide the original features and more background information about the data. Features V1, V2, … V28 are the principal components obtained with PCA, the only features which have not been transformed with PCA are 'Time' and 'Amount'. Feature 'Time' contains the seconds elapsed between each transaction and the first transaction in the dataset. The feature 'Amount' is the transaction Amount, this feature can be used for example-dependant cost-sensitive learning. Feature 'Class' is the response variable and it takes value 1 in case of fraud and 0 otherwise.

Given the class imbalance ratio, we recommend measuring the accuracy using the Area Under the Precision-Recall Curve (AUPRC). Confusion matrix accuracy is not meaningful for unbalanced classification.

### Context
It is important that credit card companies are able to recognize fraudulent credit card transactions so that customers are not charged for items that they did not purchase.

### Modelo KNN

```python
# Vamos a querer graficar los distintos valores del score de cross validation
# en función del hiperparámetro n_neighbors. Para esto generamos una lista de
# diccionarios que después se puede convertir fácilmente en DataFrame.

# Probamos todos los enteros desde el 1 hasta el 20 como posibles valores de 
# n_neighbors a explorar.

# Definimos la estrategia de validación cruzada - dejamos un semilla para que 
# sea replicable
kf = KFold(n_splits=5, shuffle=True, random_state=12)

scores_para_df = []
#lista_cant_vecinos = [1,2,3,4,5,6,7]

for i in range(1, 7):
    
    # En cada iteración, instanciamos el modelo con un hiperparámetro distinto
    model = KNeighborsClassifier(n_neighbors=i)
    
    # cross_val_scores nos devuelve un array de 5 resultados,
    # uno por cada partición que hizo automáticamente CV
    # La métrica de decisión será f1 score
    cv_scores = cross_val_score(model, Xtrain, ytrain, cv=kf, scoring='f1_macro')
    
    # Para cada valor de n_neighbours, creamos un diccionario con el valor
    # de n_neighbours y la media y el desvío de los scores
    dict_row_score = {'score_media':np.mean(cv_scores),
                      'score_std':np.std(cv_scores), 'n_neighbors':i}
    
    # Guardamos cada uno en la lista de diccionarios
    scores_para_df.append(dict_row_score)

# Creamos el DataFrame a partir de la lista de diccionarios
df_scores = pd.DataFrame(scores_para_df)
df_scores
```

```python
# Identificamos el score máximo
print(df_scores.loc[df_scores.score_media == df_scores.score_media.max()])
```

```python
# Realizamos un gráfico con el comportamiento del score elegido ( en esta caso f1 score) frente a la cantidad de vecinos

plt.plot(df_scores['n_neighbors'], df_scores['score_media'], color='b')
```

```python
# Elegimos el modelo óptimo de acuerdo a las pruebas de cross validation (identificación de número de vecinos óptimo)

knn = KNeighborsClassifier(n_neighbors=3)

# Ajustamos el modelo sobre los datos de entrenamiento
knn.fit(Xtrain, ytrain)
```

```python
# Calculamos las predicciones y métricas sobre la data de entrenamiento

predicciones=knn.predict(Xtrain)
sns.heatmap(confusion_matrix(ytrain, predicciones), annot=True, fmt='4d')
plt.xlabel('Predichos')
plt.ylabel('Reales')
plt.show()
print(classification_report(ytrain, predicciones, digits=4))
```

```python
# Calculamos las predicciones y métricas sobre la data de prueba

predicciones=knn.predict(Xtest)
sns.heatmap(confusion_matrix(ytest, predicciones), annot=True, fmt='4d')
plt.xlabel('Predichos')
plt.ylabel('Reales')
plt.show()
print(classification_report(ytest, predicciones, digits=4))
```

## Comparativa de modelos y elección del mejor modelo

### Curva ROC - AUC

```python
# Listamos los modelos que hemos entrenado

nombre_modelos = ['Regresión Logística',
          'Naive Bayes Gaussian',
          'KNN',
          'Decision Tree',
          'Random Forest']

modelos = [logistic_regression,gnb,knn,decision_tree,random_forest]

# Lista de valores de AUC para todos los modelos
lista_auc = []

# Construimos un ciclo para calcular y graficar la curva ROC-AUC usando todos los modelos entrenados
for modelo, nombre_modelo in zip(modelos, nombre_modelos):
    # Predecir las probabilidades (no se predice la clase)
    probs = modelo.predict_proba(Xtest)
    # Guardamos la probabilidad de ser prepago solamente
    probs = probs[:,1]
    # Calculamos el área bajo la curva AUC
    auc = roc_auc_score(ytest, probs)
    # Guardamos el valor de auc en la lista
    lista_auc.append(auc)
    # Calcular la curva ROC - Receiver Operating Characteristic
    fpr, tpr, thresholds = roc_curve(ytest, probs)
    # Calcular el g-mean para cada threshold
    gmeans = np.sqrt(tpr * (1-fpr)) # Raiz del producto de verdaderos positivos y verdaderos negativos
    # Localizamos el indice donde g-mean alcanza su máximo valor
    ix = np.argmax(gmeans)
    # Construir el gráfico de la curva ROC
    plt.plot([0,1], [0,1], linestyle="--", label="Sin rendimiento")
    plt.plot(fpr, tpr, marker=".", label=nombre_modelo)
    plt.scatter(fpr[ix], tpr[ix], marker="o", color="black", label="Mejor", s=100)
    # Ponemos labels a los ejes
    plt.xlabel("FPR - Tasa de Falsos Positivos")
    plt.ylabel("TPR - Tasa de Verdaderos Positivos")
    # Mostrar la leyenda
    plt.legend()
    # Mostrar el título
    plt.title("Curva ROC-AUC " + nombre_modelo)
    # Mostrar el gráfico
    plt.show()
    # Resumir scores y mejor threshold
    print("ROC AUC=%.3f" % (auc))
    print("Mejor threshold=%.3f, G-Mean=%.3f" % (thresholds[ix], gmeans[ix]))
    print("\n\n")
```

### Tabla comparativa de modelos

```python
# Calculamos los diferentes métricas para los modelos entrenados
# y generamos un resumen comparativo

lista_accuracy = []
lista_recall = []
lista_precision = []
lista_f1score = []

for modelo in modelos:
    ypred = modelo.predict(Xtest)
    lista_accuracy.append(accuracy_score(ytest, ypred))
    lista_recall.append(recall_score(ytest, ypred))
    lista_precision.append(precision_score(ytest, ypred))
    lista_f1score.append(f1_score(ytest, ypred))

comparativa_dict = {'Modelo': nombre_modelos,
                   'Accuracy': lista_accuracy,
                   'Recall': lista_recall,
                   'Precision': lista_precision,
                   'F1 score': lista_f1score,
                   'AUC': lista_auc}

comparativa = pd.DataFrame(comparativa_dict)
```

```python
# Mostramos la tabla comparativa de modelos

display(comparativa)
```

Teniendo en cuenta la tabla comparativa evidenciamos que el modelo que presenta mejor comportamiento es el **Random Forest**, siendo superior en todas las métricas con solo una excepción respecto a los otros modelos. La excepción es que el Modelo Naive Bayes Gaussian nos arroja un mejor recall, pero con un costo bastante alto el cual es la marca de una cantidad bastante alta de falsos positivos, que se traducen en una precisión y f1 score muy pobres.

De este modo, el modelo definitivo para modelar el problema de Fraude en tarjeta de crédito es el Random Forest, sobre el cual realizaremos la elección del mejor threshold para su implementación.

### Tabla comparativa de métricas para diferentes thresholds

```python
# Calculamos los diferentes métricas para los modelos entrenados
# y generamos un resumen comparativo

lista_accuracy = []
lista_recall = []
lista_precision = []
lista_f1score = []

thresholds = np.arange(0.05, 0.95,0.05)

for threshold in thresholds:
    preds = random_forest.predict_proba(Xtest)[:,1]
    ypred = np.where(preds>=threshold, 1,0)
    lista_accuracy.append(accuracy_score(ytest, ypred))
    lista_recall.append(recall_score(ytest, ypred))
    lista_precision.append(precision_score(ytest, ypred))
    lista_f1score.append(f1_score(ytest, ypred))

comparativa_dict = {'Threshold': thresholds,
                   'Accuracy': lista_accuracy,
                   'Recall': lista_recall,
                   'Precision': lista_precision,
                   'F1 score': lista_f1score}

comparativa = pd.DataFrame(comparativa_dict)
```

Observamos que el valor más alto de f1 score se obtiene con un threshold cercano a 0.05, por lo cual realizaremos un zoom en los thresholds de 0.40 a 0.60.
